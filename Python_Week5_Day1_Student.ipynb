**Matplotlib:**

Matplotlib is  used to create 2D graphs and plots by using python scripts It provides a MATLAB-like interface and supports a wide variety of plots including line plots, scatter plots, bar plots, histograms, pie charts, and more. Matplotlib is highly customizable, allowing users to fine-tune every aspect of their plots. It is one of the most widely used libraries for data visualization in Python due to its versatility and ease of use.

**Seaborn:**

Seaborn is a statistical data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn simplifies the process of creating complex visualizations such as heatmaps, violin plots, box plots, and pair plots. It also offers built-in themes and color palettes to enhance the aesthetics of your plots. Seaborn is particularly useful for visualizing statistical relationships in datasets and is often used in conjunction with Pandas data structures.









**Key Differences between seaborn and matplotlib:**

**Defaults:** Seaborn applies default styling and aesthetics, while Matplotlib requires more manual configuration.

**Convenience:**  Seaborn often simplifies plot creation with less code.

**Types of Data:**

**1. Nominal Data**

**Definition:** Data that represents categories or labels without any inherent order. Values can't be ranked or compared mathematically.

**Examples:**

Gender (Male, Female, Other)

Eye color (Blue, Brown, Green)

Country of origin (India, USA, Japan)

Types of fruit (Apple, Banana, Orange)


**2. Ordinal Data**

**Definition:** Data that has a natural order or ranking. However, the differences between the values are not necessarily equal or meaningful.

**Examples:**

Survey ratings (Strongly Disagree, Disagree, Neutral, Agree, Strongly Agree)

Educational levels (High School, Bachelor's, Master's, PhD)

Performance ranking (First, Second, Third)

Clothing sizes (Small, Medium, Large)

**3. Discrete Data**

**Definition:** Data that represents counts of whole numbers. It can only take on specific values, with nothing in between.

**Examples:**

Number of children in a family

Number of cars on the road

Number of rooms in a house

Number of defects in a product

**4. Continuous Data**

**Definition:** Data that can take on any value within a range. It can be infinitely subdivided. Measurements belong to this category.

**Examples:**

Height (e.g., 5.8 feet)

Weight (e.g., 72 kilograms)

Temperature (e.g., 22.5 degrees Celsius)

Time (e.g., 3.45 seconds)

**Scatter Plot:**

A scatter plot is a visual representation of how two variables relate to each other. You can use scatter plots to explore the relationship between two variables


**Line Plot:**

A line plot visualizes information as a series of data points called markers connected by straight line segments

**Bar Plot:**

Visualizes categories and their associated values using vertical bars.

**Pie Chart:**

Definition: Visualizes proportions of a categorical variable as slices of a pie.

Shows the distribution of a numerical variable by dividing its range into bins and counting the number of values in each bin.
