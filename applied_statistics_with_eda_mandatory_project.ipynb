Load the Dataset and plot the histogram on Netflix Content Release Year Distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

netflix_df = pd.read_csv('netflix.csv')

# Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=netflix_df, x='release_year', bins=30, color='skyblue', edgecolor='black')
plt.title('Netflix Content Release Year Distribution')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Load the Dataset and plot the Distribution of Populartiy of different content categories using countplot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
netflix_df = pd.read_csv('netflix.csv')

# Split the 'listed_in' column into individual genres and explode the list
netflix_df['listed_in'] = netflix_df['listed_in'].str.split(', ')
exploded_df = netflix_df.explode('listed_in')

# Plot countplot
plt.figure(figsize=(12, 8))
sns.countplot(data=exploded_df, y='listed_in', order=exploded_df['listed_in'].value_counts().index)
plt.title('Distribution of Popularity of Different Content Categories')
plt.xlabel('Number of Shows/Movies')
plt.ylabel('Content Category')
plt.show()
```
