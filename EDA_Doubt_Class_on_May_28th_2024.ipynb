```python
## treating missing values
# univariant imputation:--- mean, median, mode
# multivatiant imputation:-- KNN, MICE

# numerical columns:
# fill with mean when no outliers
# fill with median for outliers

# categorical columns:
# fill with mode
```

```python
#check outliers -- using box plot
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(df1['Age'])
plt.subplot(1,2,2)
sns.boxplot(df1['Family size'])
plt.show()
```

```python
# based on above graph age is having outliers will fill with median
# family size: no outliers fill with mean
age_median = df1['Age'].median()
fs_mean = df1['Family size'].mean()
```

```python
print(age_median)
print(fs_mean)
```

```python
df1['Age'].fillna(age_median, inplace=True)
df1['Family size'].fillna(fs_mean, inplace=True)
```

```python
df1.isnull().sum()
```

```python
#check outliers -- using box plot
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(df1['Age'])
plt.subplot(1,2,2)
sns.boxplot(df1['Family size'])
plt.show()
```

```python
# based on above graph, age column is having outliers, treat outliers using capping, trimming techniques
# Upper Limit = Q3+1.5*IQR
# Lower Limit = Q1-1.5*IQR
Q3 = df1['Age'].describe()['75%']
print(Q3)
Q1 = df1['Age'].describe()['25%']
print(Q1)
IQR = Q3-Q1
IQR
```

```python
UL = Q3+1.5*IQR
print(UL)
LL = Q1-1.5*IQR
print(LL)
```

```python
## Capping technique:-- fill outliers with lower limit and upper limit values
df1['Age'] = df1['Age'].clip(LL, UL)
```

```python
#check outliers -- using box plot
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(df1['Age'])
plt.subplot(1,2,2)
sns.boxplot(df1['Family size'])
plt.show()
```
