1. It has been claimed that manual cars have high mileage per gallon. Use mtcars dataset and create box plot between am and mpg columns and compare cylinders also in the same plot.

```python
import pandas as pd
import plotly.express as px

mtcars_df = pd.read_csv('mtcars - mtcars (1).csv')
mtcars_df['am'] = mtcars_df['am'].map({0: 'Automatatic', 1: 'Manual'})

fig = px.box(mtcars_df, x='am', y='mpg', color='cyl',
             title='Box Plot of Mileage per Gallon (mpg) by Transmission Type and Number of Cylinders')
fig.show()
```

2. Use mtcars dataset and create a heatmap for mpg, disp, hp, wt and qsec columns.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

mtcars_df = pd.read_csv('mtcars - mtcars (1).csv')
selected_columns = ['mpg', 'disp', 'hp', 'wt', 'qsec']
df = mtcars_df[selected_columns]

cor_matrix = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.5)
plt.title('Heatmap of Correlation Between mpg, disp, hp, wt, and qsec')
plt.show()
```

3. Use mtcars dataset and create histogram of mpg column and scatterplot between mpg and weight using subplot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

mtcars_df = pd.read_csv('mtcars - mtcars (1).csv')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,6))

sns.histplot(mtcars_df, x='mpg', ax=axes[0])
axes[0].set_title('Histogram of mpg')

sns.scatterplot(data=mtcars_df, x='mpg', y='wt', ax=axes[1])
axes[1].set_title('Scatter Plot of mpg vs weight')

plt.tight_layout()
plt.show()
```

4. Use mtcars dataset and Examine the relationship between am and vs columns using countplot also add legend at upper left.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

mtcars_df = pd.read_csv('mtcars - mtcars (1).csv')

mtcars_df['am'] = mtcars_df['am'].map({0: 'Automatic', 1: 'Manual'})
mtcars_df['vs'] = mtcars_df['vs'].map({0: 'V-Shaped', 1: 'Straight'})

plt.figure(figsize=(8,6))
sns.countplot(data=mtcars_df, x='am', hue='vs')
plt.legend(title='Engine Type', loc='upper left')
plt.title('Countplot of Transmission Type (am) vs Engine Type (vs)')
plt.xlabel('Transmission Type (am)')
plt.ylabel('Count')
plt.show()
```

5. Create a scatter plot using plotly.express. Use the following data:

    data = {'X': [1, 2, 3, 4, 5],
    'Y': [10, 12, 8, 15, 9]}

```python
import pandas as pd
import plotly.express as px

data = {'X': [1, 2, 3, 4, 5], 'Y': [10, 12, 8, 15, 9]}
df = pd.DataFrame(data)

fig = px.scatter(df, x='X', y='Y', title='Scatter Plot: X vs Y')
fig.show()
```
