Compute the correlation coefficient between 'median_income' and 'median_house_value' columns.

```python
import pandas as pd

# Read the housing data from a CSV file
df = pd.read_csv('housing.csv')

# Calculate the Pearson correlation coefficient between 'median_income' and 'median_house_value'
correlation_coefficient = df['median_income'].corr(df['median_house_value'])

# Print the correlation coefficient
print(correlation_coefficient)
```

Identify and remove outliers in the 'median_house_value' column using the IQR method.

```python
import pandas as pd

# Read the housing data from a CSV file
df = pd.read_csv('housing.csv')

# Calculate the first quartile (Q1) and third quartile (Q3) for 'median_house_value'
Q1 = df['median_house_value'].quantile(0.25)
Q3 = df['median_house_value'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers based on the bounds
outliers = df[(df['median_house_value'] < lower_bound) | (df['median_house_value'] > upper_bound)]

# Create a cleaned DataFrame by dropping the outlier rows
cleaned_df = df.drop(outliers.index)

# Print the cleaned DataFrame
print(cleaned_df)
```

Implement a class named DataVisualizer with a method to plot a histogram of 'median_house_value'.
Create an object for the class and call the method to show histogram in the output.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the housing data from a CSV file
df = pd.read_csv('housing.csv')

# Define a DataVisualizer class
class DataVisualizer:
    def __init__(self, data):
        self.data = data

    def plot_histogram(self):
        # Create a histogram of the 'median_house_value' column
        plt.hist(self.data['median_house_value'], bins=20, color='skyblue', edgecolor='black')
        plt.xlabel('Median House Value')
        plt.ylabel('Frequency')
        plt.title('Histogram of Median House Value')
        plt.grid(True)
        plt.show()

# Create an instance of DataVisualizer with the housing data
visualizer = DataVisualizer(df)

# Call the plot_histogram method to display the histogram
visualizer.plot_histogram()
```

Create a scatter plot of 'median_income' against 'median_house_value'.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the housing data from a CSV file
df = pd.read_csv('housing.csv')

# Create a scatter plot of 'median_income' against 'median_house_value'
plt.figure(figsize=(10, 6))
plt.scatter(df['median_income'], df['median_house_value'], alpha=0.5, color='b')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Scatter Plot: Median Income vs Median House Value')
plt.grid(True)
plt.show()
```
