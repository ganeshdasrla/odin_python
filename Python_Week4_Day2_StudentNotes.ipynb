##**Regular expressions:**

Regular expressions, or RegEx, are a powerful tool for defining patterns to match, search, and manipulate text.

image.png

**re.match(pattern, string):**

**Purpose:** Matches the pattern only at the beginning of the string.

**Returns:** A match object if found, otherwise None

**re.search(pattern, string):**

**Purpose:** Finds the first occurrence of the pattern in the string.

**Returns:** A match object if found, otherwise None.

**Match Object**

A Match Object is an object containing information about the search and the result.

The Match object has properties and methods used to retrieve information about the search, and the result:

**.span()** returns a tuple containing the start-, and end positions of the match.

**.string** returns the string passed into the function.

**.group()** returns the part of the string where there was a match

**re.findall(pattern, string):**

Finds all non-overlapping occurrences of a regular expression pattern within a string and returns a list of those matches.

It's a powerful tool for extracting specific text patterns from strings

##**MetaCharacters:**

#. --> it gets you all the characters except the new line

```python
string3="Google Gogle \n Ggle Gooogle."
#. it gets you all the characters except the new line
# please print all the characters from the string
matches=re.finditer(".",string3)
for i in matches:
  print(i)
```

# * -->it matches 0 or more characters

```python
string3="Google Gogle \n Ggle Google."
# * it matches 0 or more characters

matches=re.finditer("Go*gle",string3) # G gle
#print(matches)
for i in matches:
  print(i)
```

# +--> it matches 1 or more characters

```python
string3="Google Gogle \n Ggle Google."
# + it matches 1 or more characters

matches=re.finditer("Go+gle",string3) # G gle
#print(matches)
for i in matches:
  print(i.span())
```

```python
string4="My birthday is on 10th apr"

matches=re.finditer("\d+",string4) # G gle
#print(matches)
for i in matches:
  print(i)
```

```python
string4="Hi this is Odinschool's 23rd class"

matches=re.finditer("[0-9]+",string4) # G gle
#print(matches)
for i in matches:
  print(i)
```

```python
string4="Hi this is Odinschool's 23rd class"

matches=re.finditer("\D+",string4) # G gle
#print(matches)
for i in matches:
  print(i)
```

```python
string4="Hi this is Odinschool's 23rd class"

matches=re.finditer(".+",string4) # G gle
#print(matches)
for i in matches:
  print(i)
```

```python
string4="Hi this - is Odinschool's 23rd class"

matches=re.finditer("\w+",string4) # G gle
#print(matches)
for i in matches:
  print(i)
```

```python
import re

text = "Playing, singing, dancing"

# With `r` (correct):
m1= re.search(r"\w+ing\b", text)  # Matches "Playing", "singing", "dancing"

# Without `r` (incorrect):
m2 = re.search("\w+ing\b", text)  # Matches "ing" only (incorrect due to escaped backslash)
m1
```
