# Implementing Simple Linear Regression using Python

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split # takes 4 parameters, X,y, random_state=0
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_reg=LinearRegression()
```

```python
linear_reg.fit(x_train,y_train) # m and c
```

```python
linear_reg.coef_
```

```python
linear_reg.intercept_
```

```python
# Test the model
y_predicted=linear_reg.predict(x_test)
```

```python
y_predicted
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(y_test,y_predicted))
```

```python
# root mean squared error(rmse)
rmse=np.sqrt(mse)
print(rmse)
```

# Logistic Regression Implementation using Python

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
# class, alone,who, adult_male,embark_town,deck,alive

df.drop(columns=["class", "alone","who", "adult_male","embark_town","deck","alive"],inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# mean -
df["age"]=df["age"].fillna(df["age"].mean())
```

```python
df.isnull().sum()
```

```python
df["embarked"].mode()[0]
```

```python
df["embarked"]=df["embarked"].fillna(df["embarked"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
# pd.get_dummies

df=pd.get_dummies(df,columns=["sex","embarked"],drop_first=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df["sex_male"]=df["sex_male"].astype("int")
df["embarked_Q"]=df["embarked_Q"].astype("int")
df["embarked_S"]=df["embarked_S"].astype("int")
```

```python
df.head()
```

```python
X=df.drop(columns=["survived"])
y=df["survived"]
```

```python
X.head()
```

```python
# 80 and 20
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Training
# x_train,y_train
# testing
# x_test= ypred
# evaluation
# y_test
```

```python
# Training the model
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
# testing
ypred=log.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
# metrics, accuracy, confusion matrix, precision recall and f1-score
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
confusion_matrix(y_test,ypred)
```

```python
accuracy_score(y_test,ypred) # 81
```

```python
precision_score(y_test,ypred)
```

```python
recall_score(y_test,ypred)
```

```python
f1_score(y_test,ypred)
```
