```python
# Outliers detection and treatment techniques
# numrical columns:
# Detection Techniques:
# boxplot, IQR, z-score, percentile method

# Treatment:
# 1. Capping 2. Trimming
# Capping: replace outliers with lower limit and upper limit
# Trimming: remove outliers
```

```python
px.scatter(df, x='Age')
```

```python
df.describe()["Age"]
```

```python
px.box(df, y='Age')
```

```python
q1 = df.describe()["Age"]["25%"]
q3 = df.describe()["Age"]["75%"]

IQR = q3-q1
```

```python
upper_limit = q3+1.5*IQR
```

```python
upper_limit
```

```python
lower_limit = q1-1.5*IQR
```

```python
lower_limit
```

```python
# get outlier values
df[(df["Age"]<lower_limit) | (df["Age"]>upper_limit)] # Trimming - remove outliers
```

```python
# Apply capping - to apply capping we use clip method in pandas
df["Age"].clip(lower_limit, upper_limit, inplace=True)
```
