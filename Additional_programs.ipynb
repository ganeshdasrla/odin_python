```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# Load the model
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Generate an image
prompt = "A futuristic cityscape"
image = pipe(prompt).images[0]

# Save the image
image.save("generated_image.png")

# Display the image
image.show()

# Display the image in Google Colab
from IPython.display import Image as IPyImage
IPyImage("generated_image.png")
```

```python
#!pip install transformers

from transformers import pipeline

# Load a summarization model
summarizer = pipeline('summarization', model='facebook/bart-large-cnn')

# Example text
text = """
Generative AI models like GPT-3 and BERT are changing the landscape of natural language processing. They can generate human-like text, translate languages, and answer questions, among other tasks. These models are trained on massive datasets and use deep learning techniques to understand and generate text.
"""

# Generate summary
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)

print("Summary:", summary[0]['summary_text'])
```

```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import requests
from io import BytesIO

# Load the model
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
pipe = pipe.to("cuda")

# Load image from a direct URL
def load_image(image_url):
    response = requests.get(image_url)
    img = Image.open(BytesIO(response.content))
    return img

# Direct image URLs
content_image_url = "https://images.unsplash.com/photo-1574180045827-681f8a1a9622"
style_image_url = "https://images.unsplash.com/photo-1584036561566-baf8f5f1b144"

# Ensure the URLs are direct links to image files
content_image = load_image(content_image_url)
style_image = load_image(style_image_url)

# Display the images in Google Colab
content_image.show()
style_image.show()

# Save and display the images in Google Colab
content_image.save("content_image.png")
style_image.save("style_image.png")

from IPython.display import Image as IPyImage
display(IPyImage("content_image.png"))
display(IPyImage("style_image.png"))
```
