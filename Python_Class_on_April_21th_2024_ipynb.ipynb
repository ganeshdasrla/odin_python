```python
# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Atlanta'],
    'Salary': [60000, 70000, 80000, 90000, 100000]
}

df = pd.DataFrame(data)
df
```

```python
df.sample(2)
```

**df.iloc[1]** selects the row at integer position 1.

**df.iloc[[1, 3]]** selects rows at integer positions 1 and 3.

**df.iloc[1, 2]** selects the element at row 1 and column 2.

**df.iloc[1:3, 1:3]** selects a slice of rows from 1 to 2 and columns from 1 to 2.

```python
row_1 = df.iloc[1]
print("Row at position 1:")
print(row_1)

# Accessing multiple rows by integer positions (rows index 1 and 3)
rows_1_3 = df.iloc[[1, 3]]
print("\nRows at positions 1 and 3:")
print(rows_1_3)

# Accessing a single element by integer position (row index 1, column index 2)
element_1_2 = df.iloc[1, 2]
print("\nElement at row 1, column 2:")
print(element_1_2)

# Accessing a slice of rows and columns by integer positions
slice_rows_cols = df.iloc[1:3, 1:3]
print("\nSlice of rows 1 to 2 and columns 1 to 2:")
print(slice_rows_cols)
```

```python
# Creating two DataFrames
data1 = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
data2 = {'ID': [2, 3, 4], 'Age': [25, 30, 35]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Performing an inner join on 'ID'
inner_join = pd.merge(df1, df2, on='ID', how='inner')
print("Inner Join:")
print(inner_join)

# Performing a left join on 'ID'
left_join = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:")
print(left_join)

# Performing a right join on 'ID'
right_join = pd.merge(df1, df2, on='ID', how='right')
print("\nRight Join:")
print(right_join)

# Performing an outer join on 'ID'
outer_join = pd.merge(df1, df2, on='ID', how='outer')
print("\nOuter Join:")
print(outer_join)
```

```python
# Creating two DataFrames
data1 = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
data2 = {'Employee_ID': [2, 3, 4], 'Age': [25, 30, 35]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Performing an inner join on 'ID' and 'Employee_ID'
inner_join = pd.merge(df1, df2, left_on='ID', right_on='Employee_ID', how='inner')
print("Inner Join:")
print(inner_join)
```

```python
# Creating a DataFrame
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'B', 'C'],
    'Value': [10, 20, 30, 40, 50, 60, 70, 80, 90]
}

df = pd.DataFrame(data)

# Grouping the DataFrame by 'Category'
grouped = df.groupby('Category')

# Getting a specific group
group_A = grouped.get_group('A')
print(group_A)
group_B = grouped.get_group('B')
print(group_B)
```

```python
# Define a regular expression pattern
pattern = r'\b[A-Z][a-z]+\b'

# Define a sample text
text = "Hello World, this is a Sample Text with Some Capitalized Words"

# Find all occurrences of the pattern in the text
matches_iterator = re.finditer(pattern, text)

# Iterate over the match objects
for match in matches_iterator:
    print("Match found:", match.group())
    print("Start position:", match.start())
    print("End position:", match.end())
    print("Match span:", match.span())
    print()
```

```python
# Define regular expression patterns
pattern_star = r'a*b'  # Matches zero or more 'a' followed by 'b'
pattern_plus = r'a+b'  # Matches one or more 'a' followed by 'b'

# Define a sample text
text = "ab b aab aaab"

# Find all occurrences of the patterns in the text
matches_star = re.findall(pattern_star, text)
matches_plus = re.findall(pattern_plus, text)

print("Matches with *:", matches_star)
print("Matches with +:", matches_plus)
```
