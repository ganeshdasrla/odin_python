##**Lambda function:**

A lambda function  is a small anonymous function. It is called "anonymous" because it doesn't have a name like a regular function defined using the def keyword. Lambda functions can take any number of arguments, but they can only have one expression. They are often used as a way to create short, simple functions without the need for defining a full function using def.

**shape , ndim , size , len():**

**shape:** The shape attribute of a NumPy array returns a tuple representing the dimensions of the array. For example, if you have a 2D array with 3 rows and 4 columns, its shape will be (3, 4). If it's a 1D array with 5 elements, its shape will be (5,). Essentially, shape tells you the size of each dimension of the array.

**ndim:** The ndim attribute of a NumPy array returns the number of dimensions (axes) of the array. For example, a 1D array will have 1 dimension, a 2D array will have 2 dimensions, and so on.

**size:** The size attribute of a NumPy array returns the total number of elements in the array. It's the product of the elements of the shape tuple.

**len:** The len function, when applied to a NumPy array, returns the length of the first dimension of the array. For multi-dimensional arrays, len returns the size of the first dimension only.

### **Mathematical functions of Numpy:**

image.png

**np.add:** This function adds two arrays element-wise.

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.add(arr1, arr2)
print(result)
#Each element of arr1 is added to the corresponding element of arr2.
'''
output
[5 7 9]
'''
```

**np.subtract:** This function subtracts elements of one array from another element-wise.

```python
import numpy as np

arr1 = np.array([4, 5, 6])
arr2 = np.array([1, 2, 3])

result = np.subtract(arr1, arr2)
print(result)
#Each element of arr2 is subtracted from the corresponding element of arr1.

'''
output:
[3 3 3]
'''
```

**np.negative:** This function returns the negative of each element of the input array.

```python
import numpy as np

arr = np.array([1, -2, 3])

result = np.negative(arr)
print(result)

#Each element of the array is negated.
'''
#output:
[-1  2 -3]
'''
```

**np.multiply:** This function multiplies two arrays element-wise.

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

result = np.multiply(arr1, arr2)
print(result)

# Each element of arr1 is multiplied by the corresponding element of arr2.
'''
output
[ 4 10 18]
'''
```

**np.divide:** This function divides the elements of one array by another element-wise.

```python
import numpy as np

arr1 = np.array([4, 6, 8])
arr2 = np.array([2, 3, 2])

result = np.divide(arr1, arr2)
print(result)

#Each element of arr1 is divided by the corresponding element of arr2.
'''
output
[2. 2. 4.]
''
```

**np.floor_divide:** This function performs floor division element-wise.

```python
import numpy as np

arr1 = np.array([10, 15, 20])
arr2 = np.array([3, 4, 5])

result = np.floor_divide(arr1, arr2)
print(result)

#Each element of arr1 is divided by the corresponding element of arr2, and the result is rounded down to the nearest integer.

'''
output
[3 3 4]
'''
```

**np.power:** This function raises elements of an array to a specified power.

```python
import numpy as np

arr = np.array([2, 3, 4])

result = np.power(arr, 2)
print(result)
# Each element of the array is raised to the power of 2.
'''
output

[ 4  9 16]
'''
```

**np.mod:** This function computes the element-wise remainder of division.

```python
import numpy as np

arr1 = np.array([10, 11, 12])
arr2 = np.array([3, 4, 5])

result = np.mod(arr1, arr2)
print(result)
'''
output
[1 3 2]
'''
```

**np.mean:** This function computes the arithmetic mean along the specified axis

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

mean_value = np.mean(arr)
print(mean_value)

#mean of the elements in the array [1, 2, 3, 4, 5] is computed as (1+2+3+4+5) / 5 = 3.0.
'''
output
3.0
'''
```

**np.median:** This function computes the median along the specified axis.

```python
import numpy as np

arr = np.array([1, 3, 5, 2, 4])

median_value = np.median(arr)
print(median_value)
#edian of the elements in the array [1, 3, 5, 2, 4] is 3.0, which is the middle value when the array is sorted.
'''
output
3.0
'''
```

**np.mode:** NumPy does not have a built-in mode function. However, you can use scipy.stats.mode from the SciPy library to compute the mode of an array.

```python
from scipy import stats
a=np.array([1,2,3,2,2,2,5,4])
stats.mode(a)

'''
output
mode=2
''
```

**np.max:** This function returns the maximum value along the specified axis.

```python
import numpy as np

arr = np.array([1, 3, 5, 2, 4])

max_value = np.max(arr)
print(max_value)

#The maximum value in the array [1, 3, 5, 2, 4] is 5.
'''
output
5
'''
```

**np.min:** This function returns the minimum value along the specified axis.

```python
import numpy as np

arr = np.array([1, 3, 5, 2, 4])

min_value = np.min(arr)
print(min_value)
#The minimum value in the array [1, 3, 5, 2, 4] is 1.
'''
output
1
'''
```

**np.std:** This function computes the standard deviation along the specified axis.
