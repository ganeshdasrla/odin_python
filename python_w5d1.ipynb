2. Use the below two lists (Age and weight) having equal number of elements and create a scatter plot. Give a proper label and title to the figure and change the marker style to diamond, add color as cyan.

    Age = [20, 24, 28, 32, 36]
    Weight = [52, 59, 68, 72, 82]

```python
import matplotlib.pyplot as plt

Age = [20, 24, 28, 32, 36]
Weight = [52, 59, 68, 72, 82]

plt.scatter(Age, Weight)
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Age vs Weight')
plt.grid()
plt.show()
```

3. Read mtcars dataset and create a histogram of mpg (mile per gallon) feature with proper lables and title. Give it an appropriate edge and face color. Add number of bins (15).

```python
import matplotlib.pyplot as plt
import pandas as pd

mtcars = pd.read_csv('mtcars - mtcars (1).csv')
plt.hist(mtcars['mpg'], bins=15, edgecolor='black', facecolor='skyblue')
plt.xlabel('Miles per Gallon (mpg)')
plt.ylabel('Frequency')
plt.title('Histogram of MPG in mtcars Dataset')
plt.show()
```

4. Use mtcars dataset and create a density plot from mtcars dataset using weight column. Add
appropriate label and title.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

mtcars = pd.read_csv('mtcars - mtcars (1).csv')
sns.kdeplot(data=mtcars, x='wt', fill=True)
plt.xlabel('Weight')
plt.ylabel('Density')
plt.title('Density Plot of Weight in mtcars Dataset')
plt.show()
```

5. Use mtcars dataset and check the distribution of 'am' column using countplot.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

mtcars = pd.read_csv('mtcars - mtcars (1).csv')
plt.figure(figsize=(8,6))
sns.countplot(data=mtcars, x='am', hue='am', palette="Set1", legend=False)
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.title('Distribution of Transmission Type in mtcars Dataset')
plt.xticks(ticks=[0, 1], labels=['Automatic', 'Manual'])
plt.show()
```

6. Use mtcars dataset and create a pair plot between mpg, disp, hp, wt and qsec columns.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mtcars = pd.read_csv('mtcars - mtcars (1).csv')
selected_columns = ['mpg', 'disp', 'hp', 'wt', 'qsec']
sns.pairplot(data=mtcars[selected_columns])
plt.suptitle("Pair Plot of mpg, disp, hp, wt, and qsec in mtcars Dataset", y=1.02)
plt.show()
```

10. Create a pie chart using the plotly library

```python
import pandas as pd
import plotly.express as px

data = {'Category': ['Category A', 'Category B', 'Category C', 'Category D'],
'Values': [25, 30, 15, 30]}
df = pd.DataFrame(data)

fig = px.pie(df, values='Values', names='Category')
fig.update_layout(title_text='Pie Chart', title_x=0.5)
fig.show()
```
