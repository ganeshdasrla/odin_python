##**Tuples:**

A tuple  is an immutable sequence of elements, similar to lists, but with the key difference that tuples cannot be modified after creation. Tuples are defined by enclosing comma-separated values within parentheses ( ). They are commonly used to represent fixed collections of items.

image.png

image.png

**Methods of Tuples:**

**count():** Returns the number of occurrences of a specified value in the tuple.

**index():** Returns the index of the first occurrence of a specified value in the tuple.

**len():** The len() function returns the number of elements in a tuple.

**Converting List to Tuple:**

To convert a list to a tuple, you can use the tuple() constructor.

**Converting Tuple to List:**

To convert a tuple to a list, you can use the list() constructor.

##**Sets:**

A set  is an unordered collection of unique elements. It is defined by enclosing comma-separated elements within curly braces {}. Sets are mutable, meaning they can be modified after creation, but they are unordered, so the elements are not indexed.
